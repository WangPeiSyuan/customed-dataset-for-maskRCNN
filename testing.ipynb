```python
import os
import torch
import cv2
from src.utils import draw_segmentation_map, get_outputs
from torchvision import transforms
```

```python
def single_img_predict(path, img, score, model):
    path = os.path.join(path, img)
#     image = np.array(Image.open(path))
#     image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    image = cv2.imread(path)
#     cv2.imwrite(os.path.join(r'D:\maskRCNN\input', img), image)
    orig_image = image.copy()
    transform = transforms.Compose([
        transforms.ToTensor()
    ])

    # transform the image
    image = transform(image)
    # add a batch dimension
    image = image.unsqueeze(0).to(device)
    masks, boxes, labels = get_outputs(image, model, score)
    result = draw_segmentation_map(orig_image, masks, boxes, labels)
#     result = np.array(result)
    return result
```

```python
model = torch.load('model/20210709maskRCNN.pth')
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
path = r'D:/maskRCNN/train_data/test'
# count=1
for img in os.listdir(path):    
    print(img)
    result = single_img_predict(path, img, 0.7, model)
    dest = os.path.join(r'D:/maskRCNN/outputs/maskRCNN_restnet50', img)
    cv2.imwrite(dest, result)
    
```
